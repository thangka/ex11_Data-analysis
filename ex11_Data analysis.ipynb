## 2. V·∫Ω ƒë·ªì th·ªã h√†m sin,cos,tan trong kho·∫£ng -2ùúã -> 2ùúã

## 2.1. V·∫Ω ƒë·ªì th·ªã h√†m sin,cos trong kho·∫£ng -2ùúã -> 2ùúã

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100, endpoint=True)
C, S = np.cos(x), np.sin(x)
plt.plot(x, C, color="red", linewidth=3, linestyle="-")
plt.plot(x, S, linewidth=3, color='green')
plt.title('Do thi Sin(x), Cos(x)')
plt.legend(['Cos(x)','Sin(x)'])
plt.xticks([-np.pi*2, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, np.pi*2],
          [r'$-2\pi$',r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$', r'$2\pi$'])
plt.yticks([-1, 0, 1])
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

## 2.2. V·∫Ω ƒë·ªì th·ªã h√†m tan trong kho·∫£ng -2ùúã -> 2ùúã

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000 , endpoint=True)
T = np.tan(x)
plt.plot(x, T, color='green', linewidth=3 )
plt.title('Do thi Tan(x)')
plt.xticks([-np.pi*2, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, np.pi*2],
          [r'$-2\pi$',r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$', r'$2\pi$'])
plt.ylim(-4, 4)
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```
